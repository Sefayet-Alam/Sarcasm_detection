```python
from pathlib import Path
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
PROCESSED_BINARY = Path("../01_data/processed/binary")
PROCESSED_TERNARY = Path("../01_data/processed/ternary")
SPLITS = Path("../01_data/interim/splits")
SPLITS.mkdir(parents=True, exist_ok=True)
```

```python
ben_binary = pd.read_csv(PROCESSED_BINARY / "ben_sarc_binary.csv")
banglasarc_binary = pd.read_csv(PROCESSED_BINARY / "banglasarc_binary.csv")
banglasarc3_binary = pd.read_csv(PROCESSED_BINARY / "banglasarc3_binary.csv")
banglasarc3_ternary = pd.read_csv(PROCESSED_TERNARY / "banglasarc3_ternary.csv")
```

```python
def make_splits(df, label_col, random_state=42):
    train_df, temp_df = train_test_split(
        df,
        test_size=0.2,
        stratify=df[label_col],
        random_state=random_state
    )

    val_df, test_df = train_test_split(
        temp_df,
        test_size=0.5,
        stratify=temp_df[label_col],
        random_state=random_state
    )

    return train_df.copy(), val_df.copy(), test_df.copy()
```

```python
ben_train, ben_val, ben_test = make_splits(ben_binary, "label_binary")
banglasarc_train, banglasarc_val, banglasarc_test = make_splits(banglasarc_binary, "label_binary")
banglasarc3b_train, banglasarc3b_val, banglasarc3b_test = make_splits(banglasarc3_binary, "label_binary")
banglasarc3t_train, banglasarc3t_val, banglasarc3t_test = make_splits(banglasarc3_ternary, "label_ternary")
```

```python
def show_split_info(name, train_df, val_df, test_df, label_col):
    print(f"\n{name}")
    print("-" * 50)
    print("Train shape:", train_df.shape)
    print(train_df[label_col].value_counts(normalize=True).sort_index())
    print("\nVal shape:", val_df.shape)
    print(val_df[label_col].value_counts(normalize=True).sort_index())
    print("\nTest shape:", test_df.shape)
    print(test_df[label_col].value_counts(normalize=True).sort_index())

show_split_info("Ben-Sarc Binary", ben_train, ben_val, ben_test, "label_binary")
show_split_info("BanglaSarc Binary", banglasarc_train, banglasarc_val, banglasarc_test, "label_binary")
show_split_info("BanglaSarc3 Binary", banglasarc3b_train, banglasarc3b_val, banglasarc3b_test, "label_binary")
show_split_info("BanglaSarc3 Ternary", banglasarc3t_train, banglasarc3t_val, banglasarc3t_test, "label_ternary")
```

```python
def save_splits(prefix, train_df, val_df, test_df, out_dir):
    train_df.to_csv(out_dir / f"{prefix}_train.csv", index=False)
    val_df.to_csv(out_dir / f"{prefix}_val.csv", index=False)
    test_df.to_csv(out_dir / f"{prefix}_test.csv", index=False)

save_splits("ben_sarc_binary", ben_train, ben_val, ben_test, SPLITS)
save_splits("banglasarc_binary", banglasarc_train, banglasarc_val, banglasarc_test, SPLITS)
save_splits("banglasarc3_binary", banglasarc3b_train, banglasarc3b_val, banglasarc3b_test, SPLITS)
save_splits("banglasarc3_ternary", banglasarc3t_train, banglasarc3t_val, banglasarc3t_test, SPLITS)
```
