```python
import os
from pathlib import Path
import pandas as pd
```

```python
BASE = Path("../01_data/raw")

ben_path = BASE / "ben_sarc"
banglasarc_path = BASE / "banglasarc"
banglasarc3_path = BASE / "banglasarc3"

print("Ben-Sarc folder exists:", ben_path.exists())
print("BanglaSarc folder exists:", banglasarc_path.exists())
print("BanglaSarc3 folder exists:", banglasarc3_path.exists())
```

```python
def list_files(folder):
    print(f"\nFiles in: {folder}")
    for p in sorted(folder.glob("*")):
        print("-", p.name)

list_files(ben_path)
list_files(banglasarc_path)
list_files(banglasarc3_path)
```

```python
def load_table(file_path):
    suffix = file_path.suffix.lower()
    if suffix == ".csv":
        return pd.read_csv(file_path)
    elif suffix in [".xlsx", ".xls"]:
        return pd.read_excel(file_path)
    else:
        raise ValueError(f"Unsupported file type: {suffix}")
```

```python
ben_file = ben_path / "Ben-Sarc.xlsx"
banglasarc_file = banglasarc_path / "SarcasDetection.csv"
banglasarc3_file = banglasarc3_path / "BanglaSarc3.xlsx"

print(ben_file)
print(banglasarc_file)
print(banglasarc3_file)
```

```python
ben_df = load_table(ben_file)
banglasarc_df = load_table(banglasarc_file)
banglasarc3_df = load_table(banglasarc3_file)
```

```python
def audit_df(name, df):
    print("\n" + "="*60)
    print(f"DATASET: {name}")
    print("="*60)
    print("Shape:", df.shape)
    print("\nColumns:")
    print(df.columns.tolist())
    print("\nHead:")
    display(df.head())
    print("\nMissing values:")
    print(df.isnull().sum())

audit_df("Ben-Sarc", ben_df)
audit_df("BanglaSarc", banglasarc_df)
audit_df("BanglaSarc3", banglasarc3_df)
```

```python
ben_text_col = "Text"
ben_label_col = "Polarity"

banglasarc_text_col = "Comments"
banglasarc_label_col = "Label"

banglasarc3_text_col = "Bangla Comments"
banglasarc3_label_col = "Sarcasm Label"
```

```python
def show_label_distribution(name, df, label_col):
    print("\n" + "-"*50)
    print(f"{name} label distribution")
    print("-"*50)
    print(df[label_col].value_counts(dropna=False))

show_label_distribution("Ben-Sarc", ben_df, ben_label_col)
show_label_distribution("BanglaSarc", banglasarc_df, banglasarc_label_col)
show_label_distribution("BanglaSarc3", banglasarc3_df, banglasarc3_label_col)
```

```python
def show_text_missing(name, df, text_col):
    print(f"{name} missing text rows: {df[text_col].isna().sum()}")

show_text_missing("Ben-Sarc", ben_df, ben_text_col)
show_text_missing("BanglaSarc", banglasarc_df, banglasarc_text_col)
show_text_missing("BanglaSarc3", banglasarc3_df, banglasarc3_text_col)
```
