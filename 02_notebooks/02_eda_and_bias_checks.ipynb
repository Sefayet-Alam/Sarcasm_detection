```python
from pathlib import Path
import pandas as pd
```

```python
CLEAN = Path("../01_data/interim/cleaned")
BINARY_OUT = Path("../01_data/processed/binary")
TERNARY_OUT = Path("../01_data/processed/ternary")

BINARY_OUT.mkdir(parents=True, exist_ok=True)
TERNARY_OUT.mkdir(parents=True, exist_ok=True)
```

```python
ben_clean = pd.read_csv(CLEAN / "ben_sarc_clean.csv")
banglasarc_clean = pd.read_csv(CLEAN / "banglasarc_clean.csv")
banglasarc3_clean = pd.read_csv(CLEAN / "banglasarc3_clean.csv")
```

```python
print("Ben-Sarc:", ben_clean["label"].value_counts(dropna=False).to_dict())
print("BanglaSarc:", banglasarc_clean["label"].value_counts(dropna=False).to_dict())
print("BanglaSarc3:", banglasarc3_clean["label"].value_counts(dropna=False).to_dict())
```

```python
# Ben-Sarc already binary
ben_binary = ben_clean.copy()
ben_binary["label_binary"] = ben_binary["label"].astype(int)

# BanglaSarc already binary
banglasarc_binary = banglasarc_clean.copy()
banglasarc_binary["label_binary"] = banglasarc_binary["label"].astype(int)

# BanglaSarc3 -> binary by dropping Neutral
banglasarc3_binary = banglasarc3_clean[
    banglasarc3_clean["label"].isin(["Sarcastic", "Non-Sarcastic"])
].copy()

banglasarc3_binary["label_binary"] = banglasarc3_binary["label"].map({
    "Non-Sarcastic": 0,
    "Sarcastic": 1
})
```

```python
banglasarc3_ternary = banglasarc3_clean.copy()

banglasarc3_ternary["label_ternary"] = banglasarc3_ternary["label"].map({
    "Non-Sarcastic": 0,
    "Neutral": 1,
    "Sarcastic": 2
})
```

```python
ben_binary = ben_binary[["text", "label_binary", "label_original", "dataset_name"]]
banglasarc_binary = banglasarc_binary[["text", "label_binary", "label_original", "dataset_name"]]
banglasarc3_binary = banglasarc3_binary[["text", "label_binary", "label_original", "dataset_name"]]

banglasarc3_ternary = banglasarc3_ternary[["text", "label_ternary", "label_original", "dataset_name"]]
```

```python
print("Ben-Sarc binary:")
print(ben_binary["label_binary"].value_counts(dropna=False))

print("\nBanglaSarc binary:")
print(banglasarc_binary["label_binary"].value_counts(dropna=False))

print("\nBanglaSarc3 binary:")
print(banglasarc3_binary["label_binary"].value_counts(dropna=False))

print("\nBanglaSarc3 ternary:")
print(banglasarc3_ternary["label_ternary"].value_counts(dropna=False))
```

```python
ben_binary.to_csv(BINARY_OUT / "ben_sarc_binary.csv", index=False)
banglasarc_binary.to_csv(BINARY_OUT / "banglasarc_binary.csv", index=False)
banglasarc3_binary.to_csv(BINARY_OUT / "banglasarc3_binary.csv", index=False)

banglasarc3_ternary.to_csv(TERNARY_OUT / "banglasarc3_ternary.csv", index=False)

print("Binary files:", sorted([p.name for p in BINARY_OUT.glob("*.csv")]))
print("Ternary files:", sorted([p.name for p in TERNARY_OUT.glob("*.csv")]))
```
